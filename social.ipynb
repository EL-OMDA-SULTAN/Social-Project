```python
import networkx as nx
```

```python
G = nx.read_edgelist('socfb-Mich67.txt', nodetype=int)
print(nx.info(G))
nx.draw(G,with_labels=True,node_color='blue',node_size=1600,font_color='white',font_size=16,)
```

```python
short_path = []
for node1 in G.nodes():
    for node2 in G.nodes():
        if(nx.has_path(G, node1, node2)==True):
            short_path.append(nx.shortest_path_length(G, node1, node2))
short_path
```
